```python
import numpy as np
P=np.array([[0.5,0.5],#Cold
            [0.4,0.6]])#Hot
B=np.array([[0.5,0.4,0.1],
            [0.2,0.4,0.4]])
m=np.array([0.2,0.8])
obs=np.array([1,3,3,2,3,3,3,2,1,2,1,1,3,1,3,3,3])-1
```

```python
a=B[:,obs[0]]*m
aind=1
while(aind<len(obs)):
    a=np.matmul(a*B[:,obs[aind]],P)
    aind=aind+1
```
